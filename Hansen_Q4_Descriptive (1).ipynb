```python
# Strip leading and trailing spaces from column names
df.columns = df.columns.str.strip()

# List of questions
questions = [
    'On average, how often do you use smart online initiatives (from Section 2) provided by the City of Coquitlam (e.g., public transportation apps, utility management apps, public service apps)?',
    'To what extent do you agree with the following statement: "Smart online initiatives provided by the City of Coquitlam have improved my overall quality of life."',
    'How easy do you find it to use the smart online initiatives provided by the City of Coquitlam?',
    'To what extent do online initiatives provided by the City of Coquitlam help you in managing your daily activities (e.g., commuting, paying bills, accessing city services)?',
    'How satisfied are you with the services provided by the online initiatives offered by the City of Coquitlam?'
]

# Generate a bar chart for each question
for question in questions:
    plt.figure(figsize=(10, 6))
    
    # Get value counts
    counts = df[question].value_counts()
    
    # Get the index of the maximum count
    max_idx = counts.idxmax()
    
    # Set bar colors
    colors = ['skyblue' if idx != max_idx else 'orange' for idx in counts.index]
    
    # Plot bar chart
    bars = counts.plot(kind='bar', color=colors)
    
    # Add labels
    for bar in bars.patches:
        bars.annotate(
            format(bar.get_height(), ',d'),
            (bar.get_x() + bar.get_width() / 2, bar.get_height()),
            ha='center', va='center',
            size=12, xytext=(0, 8),
            textcoords='offset points'
        )
    
    plt.title(question)
    plt.xlabel('Response')
    plt.ylabel('Count')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()
```

```python
total = df['Age'].size
age_weights = {
  '<25': 0.27 / (l25/total),
  '25 - 44': 0.276/(b2544/total),
  '45 - 64': 0.289 / (b4564/total),
  '>64': 0.163/(g64/total)
}
print(age_weights)
df['age_weight'] = df['Age'].map(age_weights)
print(df)
```

```python
# List of questions
questions = [
    'On average, how often do you use smart online initiatives (from Section 2) provided by the City of Coquitlam (e.g., public transportation apps, utility management apps, public service apps)?',
    'To what extent do you agree with the following statement: "Smart online initiatives provided by the City of Coquitlam have improved my overall quality of life."',
    'How easy do you find it to use the smart online initiatives provided by the City of Coquitlam?',
    'To what extent do online initiatives provided by the City of Coquitlam help you in managing your daily activities (e.g., commuting, paying bills, accessing city services)?',
    'How satisfied are you with the services provided by the online initiatives offered by the City of Coquitlam?'
]

# Function to plot weighted frequencies
def plot_weighted_frequencies(question):
    # Rename column for easier access
    df_renamed = df.rename(columns={question: 'response'})
    
    # Calculate weighted frequencies
    weighted_freq = df_renamed.groupby('response')['gender_weight'].sum().sort_values(ascending=False)
    
    # Plot pie chart
    plt.figure(figsize=(10, 8))
    plt.pie(weighted_freq.values, labels=weighted_freq.index, autopct='%1.1f%%', startangle=90)
    plt.title(f'Weighted Frequencies for: {question}')
    plt.axis('equal')
    plt.show()
    
    # Print weighted frequencies for verification
    print(weighted_freq)

# Plot weighted frequencies for each question
for question in questions:
    plot_weighted_frequencies(question)
```

```python
# List of questions
questions = [
    'On average, how often do you use smart online initiatives (from Section 2) provided by the City of Coquitlam (e.g., public transportation apps, utility management apps, public service apps)?',
    'To what extent do you agree with the following statement: "Smart online initiatives provided by the City of Coquitlam have improved my overall quality of life."',
    'How easy do you find it to use the smart online initiatives provided by the City of Coquitlam?',
    'To what extent do online initiatives provided by the City of Coquitlam help you in managing your daily activities (e.g., commuting, paying bills, accessing city services)?',
    'How satisfied are you with the services provided by the online initiatives offered by the City of Coquitlam?'
]

# Function to plot age-weighted frequencies
def plot_age_weighted_frequencies(question):
    # Rename column for easier access
    df_renamed = df.rename(columns={question: 'response'})
    
    # Calculate age-weighted frequencies
    weighted_freq = df_renamed.groupby('response')['age_weight'].sum().sort_values(ascending=False)
    
    # Plot pie chart
    plt.figure(figsize=(10, 8))
    plt.pie(weighted_freq.values, labels=weighted_freq.index, autopct='%1.1f%%', startangle=90)
    plt.title(f'Weighted Frequencies for: {question} (Age Weighted)')
    plt.axis('equal')
    plt.show()
    
    # Print weighted frequencies for verification
    print(weighted_freq)

# Plot age-weighted frequencies for each question
for question in questions:
    plot_age_weighted_frequencies(question)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

def perform_regression_analysis(question):
    # Rename column for easier access
    # df_renamed = df.rename(columns={question: 'response'})
    # print(df_renamed[question])

    response_map = {
        'Never': 1, 'Rarely': 2, 'Daily': 3, 'Weekly': 4, 'Monthly': 5,
        'Strongly disagree': 1, 'Disagree': 2, 'Neutral': 3, 'Agree': 4, 'Strongly Agree': 5,
        'Very difficult': 1, 'Difficult': 2, 'Neutral': 3, 'Easy': 4, 'Very easy': 5,
        'Very dissatisfied': 1, 'Dissatisfied': 2, 'Neutral': 3, 'Satisfied': 4, 'Very satisfied': 5
    }
    
    df['response_numeric'] = df[question].map(response_map)
    # print(df['response_numeric'])
    df['age_weight'] = pd.to_numeric(df['age_weight'], errors='coerce')
    # print(df['age_weight'])
    df_clean = df.dropna(subset=['age_weight', 'response_numeric'])
   # print(df_clean)
    # Check if we have enough data points
    if len(df_clean) < 2:
        print(f"Not enough valid data points for question: {question}")
        return
    
    # Perform regression
    slope, intercept, r_value, p_value, std_err = stats.linregress(df_clean['age_weight'], df_clean['response_numeric'])
    
    # Create scatter plot
    plt.figure(figsize=(10, 6))
    plt.scatter(df_clean['age_weight'], df_clean['response_numeric'], alpha=0.5)
    
    # Add regression line
    x = np.array([df_clean['age_weight'].min(), df_clean['age_weight'].max()])
    y = slope * x + intercept
    plt.plot(x, y, color='red', label=f'y = {slope:.2f}x + {intercept:.2f}')
    
    plt.xlabel('Age')
    plt.ylabel('Response')
    plt.title(f'Regression Analysis for: {question}')
    plt.legend()
    
    # Add R-squared and p-value to the plot
    plt.text(0.05, 0.95, f'R-squared: {r_value**2:.3f}', transform=plt.gca().transAxes)
    plt.text(0.05, 0.90, f'p-value: {p_value:.3f}', transform=plt.gca().transAxes)
    
    plt.show()
    
    print(f"Regression results for: {question}")
    print(f"Slope: {slope:.3f}")
    print(f"Intercept: {intercept:.3f}")
    print(f"R-squared: {r_value**2:.3f}")
    print(f"p-value: {p_value:.3f}")
    print(f"Number of valid data points: {len(df_clean)}")
    print("\n")

# Perform regression analysis for each question
for question in questions:
    perform_regression_analysis(question)
```
