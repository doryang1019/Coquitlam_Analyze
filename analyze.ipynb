```python
initiative_aliases = {
    'Tracking and receiving reminders for curbside collection with the Coquitlam Curbside Collection App.': [],
    'Paying utility bills through eService': [],
    'Viewing active, open, and closed projects on the Let\'s Talk Coquitlam website.': [],
    'Searching for jobs in eService through the Coquitlam careers portal at https://careers.coquitlam.ca/': [],
    'Accessing free Wi-Fi in city parks or civic facilities.': [],
    'Renewing dog licenses through the eService website.': [],
    'Donate shelter via the Coquitlam eService Online donation form': [],
    'Viewing bill histories, access property tax, utilities, and manage dog license accounts through the myCoquitlam website.': [],
    'None of the above.': []
}

# Function to count initiatives
def count_initiatives(series):
    counts = {initiative: 0 for initiative in initiative_aliases.keys()}
    for response in series:
        if isinstance(response, str):
            for initiative in initiative_aliases.keys():
                if initiative in response:
                    counts[initiative] += 1
    return pd.Series(counts)

# Function to add value labels on bars
def add_value_labels(ax, spacing=5):
    for rect in ax.patches:
        y_value = rect.get_height()
        x_value = rect.get_x() + rect.get_width() / 2
        label = f"{int(y_value)}"
        ax.annotate(label, (x_value, y_value), xytext=(0, spacing),
                    textcoords="offset points", ha='center', va='bottom')

# # Apply gender weights
# gender_weights = {
#     'Female': 0.509 / (199/393),
#     'Male': 0.491 / (194/393)
# }
# df['gender_weight'] = df['Gender'].map(gender_weights)

# Count usage of initiatives
initiative_column = 'Have you ever used any of the following smart online initiatives implemented by Coquitlam?'
usage_counts = count_initiatives(df[initiative_column])

# Plotting initiative usage (Bar Chart)
plt.figure(figsize=(12, 6))
ax = usage_counts.plot(kind='bar')
plt.title('Usage of Smart Initiatives in Coquitlam')
plt.xlabel('Initiative')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
add_value_labels(ax)
plt.tight_layout()
plt.show()
plt.savefig('initiative_usage_bar.png')
plt.close()

# Plotting initiative usage (Pie Chart)
plt.figure(figsize=(10, 10))
plt.pie(usage_counts, labels=usage_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Usage of Smart Initiatives in Coquitlam')
plt.axis('equal')
plt.tight_layout()
plt.show()
plt.savefig('initiative_usage_pie.png')
plt.close()



print("Analysis complete. Plots saved as PNG files.")
```

```python
import pandas as pd
import scipy.stats as stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
satisfaction_mapping = {
    'Very easy': 5,
    'Easy': 4,
    'Neutral': 3,
    'Difficult': 2,
    'Very difficult': 1,
    'Never used before': 0  # or you might want to treat this as a separate category
}
df['my_coquitlam_score'] = df['myCoquitlam_satisfaction'].dropna().map(satisfaction_mapping)
print(df['my_coquitlam_score'])
# Perform weighted one-way ANOVA
model = ols('my_coquitlam_score ~ C(myCoquitlam_satisfaction)', 
            data=df, 
            weights=df['age_weight']).fit()

anova_table = sm.stats.anova_lm(model, typ=2)
model2 = ols('my_coquitlam_score ~(myCoquitlam_satisfaction)', data = df, weights = df['gender_weight']).fit()
anova_table2 = sm.stats.anova_lm(model2, typ=2)
print(anova_table)
print(anova_table2)
```

```python
import pandas as pd
import numpy as np
from statsmodels.stats.weightstats import DescrStatsW
from scipy import stats


df_clean['my_coquitlam_score'] = pd.to_numeric(df_clean['my_coquitlam_score'], errors='coerce')


df_clean = df_clean.dropna(subset=['Age', 'my_coquitlam_score', 'age_weight'])

#weighted one-way ANOVA
age_groups = df_clean['Age'].unique()
group_data = [df_clean[df_clean['Age'] == age]['my_coquitlam_score'] for age in age_groups]
group_weights = [df_clean[df_clean['Age'] == age]['age_weight'] for age in age_groups]

f_statistic, p_value = stats.f_oneway(*[DescrStatsW(data, weights=weights).data for data, weights in zip(group_data, group_weights)])

print("One-way ANOVA results:")
print(f"F-statistic: {f_statistic}")
print(f"p-value: {p_value}")

# If you want to see the means for each group
for age, data, weights in zip(age_groups, group_data, group_weights):
    weighted_mean = np.average(data, weights=weights)
    print(f"Weighted mean for {age}: {weighted_mean}")
```
