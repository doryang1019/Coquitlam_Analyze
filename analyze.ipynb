```python
import pandas as pd
import scipy.stats as stats
import statsmodels.api as sm
from statsmodels.formula.api import ols
satisfaction_mapping = {
    'Very easy': 5,
    'Easy': 4,
    'Neutral': 3,
    'Difficult': 2,
    'Very difficult': 1,
    'Never used before': 0  # or you might want to treat this as a separate category
}
df['my_coquitlam_score'] = df['myCoquitlam_satisfaction'].dropna().map(satisfaction_mapping)
print(df['my_coquitlam_score'])
# Perform weighted one-way ANOVA
model = ols('my_coquitlam_score ~ C(myCoquitlam_satisfaction)', 
            data=df, 
            weights=df['age_weight']).fit()

anova_table = sm.stats.anova_lm(model, typ=2)
model2 = ols('my_coquitlam_score ~(myCoquitlam_satisfaction)', data = df, weights = df['gender_weight']).fit()
anova_table2 = sm.stats.anova_lm(model2, typ=2)
print(anova_table)
print(anova_table2)
```

```python
from scipy.stats import chi2_contingency

contingency_table = pd.crosstab(df['Age'], df['my_coquitlam_score'])
chi2, p_value, dof, expected = chi2_contingency(contingency_table)

print(f"Chi-square statistic: {chi2}")
print(f"p-value: {p_value}")
```

```python
import pandas as pd
import numpy as np
from statsmodels.stats.weightstats import DescrStatsW
from scipy import stats


df_clean['my_coquitlam_score'] = pd.to_numeric(df_clean['my_coquitlam_score'], errors='coerce')


df_clean = df_clean.dropna(subset=['Age', 'my_coquitlam_score', 'age_weight'])

#weighted one-way ANOVA
age_groups = df_clean['Age'].unique()
group_data = [df_clean[df_clean['Age'] == age]['my_coquitlam_score'] for age in age_groups]
group_weights = [df_clean[df_clean['Age'] == age]['age_weight'] for age in age_groups]

f_statistic, p_value = stats.f_oneway(*[DescrStatsW(data, weights=weights).data for data, weights in zip(group_data, group_weights)])

print("One-way ANOVA results:")
print(f"F-statistic: {f_statistic}")
print(f"p-value: {p_value}")

# If you want to see the means for each group
for age, data, weights in zip(age_groups, group_data, group_weights):
    weighted_mean = np.average(data, weights=weights)
    print(f"Weighted mean for {age}: {weighted_mean}")
```

```python
import pandas as pd
import numpy as np

# Assuming df_clean is your DataFrame
s = """On a scale of 1 to 5, how concerned are you about the following privacy issues related to smart applications?  
(1 = Not concerned at all, 5 = Extremely concerned)  [Unauthorized access to personal data]"""

# Function to calculate weighted median
def weighted_median(data, weights):
    sorted_data = sorted(zip(data, weights))
    cumulative_weight = 0
    total_weight = sum(weights)
    for data_point, weight in sorted_data:
        cumulative_weight += weight
        if cumulative_weight >= total_weight / 2:
            return data_point

# Calculate weighted statistics
stats = df_clean.groupby(s).apply(lambda x: pd.Series({
    'weighted_median': weighted_median(x[s], x['age_weight']),
    'weighted_mean': np.average(x[s], weights=x['age_weight']),
    'max': x[s].max(),
    'min': x[s].min(),
    'count': x['age_weight'].sum()  # This gives the weighted count
}))

# Sort the index to match the 1-5 scale
stats = stats.sort_index()

print(stats)

# Calculate overall weighted statistics
overall_stats = pd.Series({
    'weighted_median': weighted_median(df_clean[s], df_clean['age_weight']),
    'weighted_mean': np.average(df_clean[s], weights=df_clean['age_weight']),
    'max': df_clean[s].max(),
    'min': df_clean[s].min(),
    'total_weight': df_clean['age_weight'].sum()
})

print("\nOverall Weighted Statistics:")
print(overall_stats)
```
